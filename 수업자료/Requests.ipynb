```python
info = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36'}
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
```

```python
print(html_src)
```

```python
soup = BeautifulSoup(html_src,'lxml')
print(soup)
```

```python
srclist = soup.select('.checkEllipsis')
len(srclist)
```

```python
song = [item.text for item in soup.select('.ellipsis.rank01 a')]
len(song)
```

```python
name = [item.text for item in soup.select('.checkEllipsis')]
len(name)
```

```python
songDF = pd.DataFrame({"노래제목":song, "가수":name})
songDF.index=songDF.index+1
songDF
```

```python
songDF.to_excel('melon100.xlsx', index=False)
```
